```python
# 读取数据
import torch
from torch.autograd import Variable
import numpy as np
import torch.nn as nn
```

```python
from torch.utils.data import Dataset,DataLoader
```

```python
class DiabetesDataset(Dataset):
    # 初始化数据 比如 下载 
    def __init__(self):
        xy = np.loadtxt('data-diabetes.csv',delimiter=',',dtype=np.float32)
        self.len = xy.shape[0]
        self.x_data = torch.from_numpy(xy[:,:-1])
        self.y_data = torch.from_numpy(xy[:,[-1]])
    def __getitem__(self,index):
        return self.x_data[index],self.y_data[index]
    def __len__(self):
        return self.len
```

```python
dataset = DiabetesDataset()
```

```python
train_data = DataLoader(dataset=dataset,batch_size=32,shuffle=True,num_workers=2)
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.linear1 = nn.Linear(8,6)
        self.linear2 = nn.Linear(6,4)
        self.linear3 = nn.Linear(4,1)
        self.sigmoid = nn.Sigmoid()
    def forward(self,x):
        out = self.sigmoid(self.linear1(x))
        out = self.sigmoid(self.linear2(out))
        out = self.sigmoid(self.linear3(out))
        return out
net = Net()
loss_fn = nn.BCELoss(size_average=True)
optimizer = torch.optim.Adam(net.parameters(),lr=0.001)
```

```python
for t in range(2):
    for i,data in enumerate(train_data):
        x_train,y_train = data
        x_train,y_train = Variable(x_train),Variable(y_train)
      
        y_pred = net(x_train)
        loss = loss_fn(y_pred,y_train)
        print(t,i,loss.data[0])
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
```
